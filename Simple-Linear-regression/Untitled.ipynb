## Simple linear Regression

****In simple linear regression we predict the value of one variable Y based on anothe variable X***

****X is called the independent variable and y is called the dependent variable***

****Why we called it simple? because it examines relationship between two variable only***

****Why it is linear? because whenthe indepedent varaible increases (or decreases) the dependent varaibale also increases(or decreses ) in linear fashion***

**** Here our Goal is to obtain a relationship (model) between outside air temperature and ice cream sales revenue***

****Once the coefficients m and b are obtained, you have obtained a simple linear regression model!***

****This “trained” model can be later used to predict any Revenue (dollars) based on the outside air Temperature!***
